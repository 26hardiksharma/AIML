```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 70
sigma = 10

x = np.linspace(30,110,1000)

y = norm.pdf(x,mu,sigma)

plt.plot(x,y)
x1 = np.linspace(mu-2*sigma,mu+2*sigma,1000)
plt.fill_between(x1,norm.pdf(x1,mu,sigma),color="green")
```
