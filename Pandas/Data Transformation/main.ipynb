```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df["tax"] = df["income"].apply(lambda x: 20 if x>= 50000 else 10)
```

```python
df
```

```python
gender_map = {
    "Male":"M",
    "Female":"F",
    "Unknown":"M"
}
```

```python
df["gender"] = df["gender"].map(gender_map)
```

```python
df
```

```python
df = df.assign(new_income = df["income"] + df["income"]*df["tax"]/100)
df
```

```python
df.columns = [i.capitalize() for i in df.columns]
df
```

```python
df.rename(columns={
    "Income":"Salary"
})
df
```

# Grouping and Aggregration

```python

df.groupby("Country")["Income"].mean()
```

```python
df.groupby("Country")["Income"].min()
```

```python
df.groupby("Country")["Income"].agg(["mean","min","max"])
```

```python
df.groupby("Country").agg({
    "Income":"mean",
    "Age":"max"
})
```

```python
# Concatenation

df1 = pd.DataFrame({
    "id":[1,2,3],
    "Name": ["Adam","Eve","Satan"]
})

df2 = pd.DataFrame({
    "id":[4,5,6],
    "Name":["Charlie","Alpha","Bravo"]
})

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
