```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X,Y = make_moons(
    n_samples=500,
    noise=0.1,
    random_state=42
)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1])
```

```python
"""
LOW Dimensions (<10 Features): 0.5-1
MID Dimensions (10-20 Features): 1-2
HIGH Dimensions (>20 Features): >2
"""

model = DBSCAN(
    eps = 0.18,
    min_samples= 5 # Around 2*amount of features as a rule of thumb.
)

labels = model.fit_predict(X_scaled)


sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1],c=labels)
```
