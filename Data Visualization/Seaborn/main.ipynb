```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style = "darkgrid")
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.relplot(tips,x="total_bill",y="tip")
```

```python
sns.relplot(tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
sns.relplot(tips,
    x="total_bill",
    y="tip",
    kind="line"
)
```

```python
sns.scatterplot(
    data = tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
sns.lineplot(
    data = tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.lineplot(
    flights,
    x="year",
    y="passengers",
)

"""
Shaded area shows confidence interval
"""
```

```python
sns.barplot(
    flights,
    x = "year",
    y="passengers"
)
```

```python
sns.boxplot(
    tips,
    x="day",
    y="tip"
)
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.histplot(
    penguins,
    x="body_mass_g",
    bins=50

)
```

```python
flights_pivot=flights.pivot(index="month",columns="year",values="passengers")
flights_pivot
```

```python
sns.heatmap(
    data=flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"  
)
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
sns.lineplot(
    tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker = "o",
    ax=ax
)

ax.set_title("Average Bill")
ax.set_xlabel("Days")
ax.set_ylabel("Total bill")
fig.tight_layout()
```
