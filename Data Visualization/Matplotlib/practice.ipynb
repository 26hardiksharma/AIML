```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
days = ["Mon","Tue","Wed","Thu","Fri"]

cities = ["New York","London","Delhi","Tokyo"]

temperatures = [
    [22,23,21,24,25],
    [18,19,17,20,21],
    [30,32,31,33,34],
    [25,26,24,27,28]
]
city = 0
fig,ax = plt.subplots(2,2);
for x in range(0,2):
    for y in range(0,2):
        ax[x][y].plot(days, temperatures[city],marker = "o")
        ax[x][y].set_title(cities[city])
        city+=1
        ax[x][y].grid(True)
fig.tight_layout()
```
