```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008,2009,2010,2011,2012]
revenue = [1005,170,427,133,232]

```

```python
# plt.plot(years,revenue)
# plt.title("Movies ")
# plt.xlabel("Year")
# plt.ylabel("Revenue")
```

```python
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378,2788,829,185,275]

plt.plot(years,revenue,label = "Oscar Movies")
plt.plot(years,non_oscar_revenue,label = "Non oscar movies")
plt.title("Oscar Movies VS Non oscar movies")
plt.legend()
plt.style.available
plt.style.use("dark_background")
```

```python
plt.bar(years,revenue)
plt.bar(years,non_oscar_revenue)
```
